```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Plot a basic line chart using plt.plot()


```python
x = [1, 2, 3, 4, 5]
y = [50, 51, 52, 48, 47]

plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x, y, color='green', linewidth=5, linestyle='-')
plt.show()
```

```python
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x, y, color='red', linewidth=5, linestyle='-')
plt.show()
```

```python
plt.plot(x, y, color='red', marker='*', linestyle='')
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
max_t = [50, 51, 52, 48, 47]
min_t = [43, 42, 40, 44, 33]
avg_t = [45, 48, 48, 46, 40]
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.show()
```

```python
plt.plot(days, max_t,label="max")
plt.plot(days, min_t,label="min")
plt.plot(days, avg_t,label="average")
plt.legend()
plt.show()
```

```python
plt.savefig('img.png', transparent=None, facecolor='auto', edgecolor='black')
plt.plot(days, max_t,label="max")
plt.plot(days, min_t,label="min")
plt.plot(days, avg_t,label="average")
plt.show()
```

```python
department = ['Hr', 'IT', 'Sales', 'Marketing']
head_count = [20, 80, 140, 80]

xpos = np.arange(len(department))
xpos
```

```python
blood_suger = [110, 85, 95, 150, 139, 88, 93, 125, 135, 80, 67, 72, 125]
plt.hist(blood_suger, rwidth=0.8)
plt.show()
```

```python
labels = ['Mango', 'Apple', 'Cherry', 'Dates']
sizes = [30, 25, 25, 20] 
colors = ['red', 'blue', 'orange', 'brown']
explode = (0.1, 0, 0, 0)  

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Fruit Distribution')
plt.axis('equal')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 10, 18, 25]
y2 = [23, 15, 12, 4, 2]

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r-o')
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.plot(x, y1, 'g--s')
plt.title('Plot 2')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, 'g-o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Custom Figure Size')
plt.show()
```

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='green', marker='x')

plt.title("Sample Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.scatter(x, y, color='red', marker='o')

for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

plt.title("Scatter Plot with Annotations")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
labels = ['R', 'P', 'G', 'S']
group1 = [10, 15, 20, 25]
group2 = [5, 10, 5, 15]

plt.bar(labels, group1, label='Group 1')
plt.bar(labels, group2, bottom=group1, label='Group 2')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 5, 6], color='orange')

ax = plt.gca()
ax.set_facecolor('black')
ax.tick_params(colors='white')
plt.title("Simple Plot", color='white')
plt.show()
```

```python
x = [1, 2, 3]
y1 = [1, 4, 9]
y2 = [9, 4, 1]

ax1 = plt.gca()
ax1.plot(x, y1, 'r')
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b')

plt.show()
```
